```python
''' Handling missing values (Intermediate).
Given a DataFrame `df` containing several missing values (`NaN` entries) in a column
named `'Score'`, write code to: (a) Drop all rows where `'Score'` is missing'''
import pandas as pd
import numpy as np

df = pd.read_csv("Gamescore.csv")
df_dropped = df.dropna(subset=["Score"])
print(df_dropped)
```

```python

''' (b) In a
separate snippet, fill those missing values with the average (mean) value of that column
instead.'''
import pandas as pd
import numpy as np

df = pd.read_csv("Gamescore.csv")
#Fill missing values with the mean mileage i.e 18.857143
df["Score"] = df["Score"].fillna(df["Score"].mean())

print(df)
```

```python
''' Grouping data and aggregating values (Intermediate).
You have a sales DataFrame `df` with columns `['Store', 'Product', 'Sales']`. Write code to
group the data by the `'Store'` column and calculate the total (sum) sales for each store.'''
import pandas as pd

df = pd.DataFrame({
    "Store": ["A", "A", "B", "B", "C"],
    "Product": ["Pen", "Book", "Pen", "Book", "Pen"],
    "Sales": [100, 200, 150, 250, 300]
})

total_sales = df.groupby("Store")["Sales"].sum()

print(total_sales)
```

```python
''' Merging two distinct DataFrames (Intermediate).
Given two DataFrames, `df_employees` (with columns `['Emp_ID', 'Name']`) and `df_salaries` (with columns `['Emp_ID', 'Salary']`).
 Write a Pandas statement to perform an inner join/merge on these DataFrames using their common column `'Emp_ID'`. '''
import pandas as pd

df_employees = pd.DataFrame({
    "Emp_ID": [101, 102, 103],
    "Name": ["Aarav", "Ananya", "Vihaan"]
})

df_salaries = pd.DataFrame({
    "Emp_ID": [101, 102, 104],
    "Salary": [50000, 60000, 55000]
})

merged_df = pd.merge(df_employees, df_salaries, on="Emp_ID", how="inner")

print(merged_df)
```

```python
''' Conditional column creation using vectorization (Intermediate).
Given a DataFrame `df` containing a numerical column `'Final_Score'`, write a statement
using Pandas or NumPy's conditional logic to create a brand new column named `'Status'`.
If `'Final_Score'` is 50 or above, the status should be `'Pass'`; otherwise, it should be `'Fail'''
import pandas as pd
import numpy as np

df = pd.DataFrame({
    "Final_Score": [75, 45, 60, 30, 50]
})

df["Status"] = np.where(df["Final_Score"] >= 50, "Pass", "Fail")

print(df)
```

```python
'''Setting and resetting structural indices (Intermediate).
Given a DataFrame `df` with a column named `'Employee_ID'`, convert this column to serve
as the structural row index of the DataFrame. Then, write a subsequent line of code that
completely reverses this action, returning the index back into a regular column.'''
import pandas as pd

df = pd.DataFrame({
    "Employee_ID": [101, 102, 103],
    "Name": ["Aarav", "Ananya", "Vihaan"]
})

# Set index
df = df.set_index("Employee_ID")
print(df)

# Reset index
df = df.reset_index()
print(df)
```
