```python
# Two sum
#make an array
l = []
n= int(input("enter number of elements in array:"))
for i in range(0,n):
    a= int(input("enter elements in array:"))
    l.append(a)
print(l)
target = int(input("enter target to reach:"))
for i in range(len(l)):
    for j in range(i,len(l)):
        if l[i]+l[j]==target and i!=j:
            print([i,j])
```

```python
# make a function to check number whether if it is prime or not
def prime(num):
  for i in range(2,num):
    if num%i==0:
      print("not prime")
      break
  else:
    print("prime")

num=int(input("enter a number: "))
prime(num)
```

```python
# make a function to check which number will be largest based on inputs
def largest(arr):
  largest=arr[0]
  for i in range(1,len(arr)):
    if arr[i]>largest:
      largest=arr[i]
  return largest

arr=[]
n= int(input("enter number of elements in array:"))
for i in range(0,n):
    a= int(input("enter elements in array:"))
    arr.append(a)
print(arr)
largest(arr)
```
